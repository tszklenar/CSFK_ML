```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((512,512))

plt.imshow(img,cmap='binary')
```

```python
# Drawing a circle
for x in range(-150,151,1):
    r = 150
    y = np.sqrt( r**2 - (x)**2 )
    img[int(y)+250,x+250] = 1
    img[-int(y)+250,x+250] = 1
```

```python
plt.imshow(img,cmap='binary',origin="lower")
```

```python
def rebin(arr, new_shape):
    """Rebin 2D array arr to shape new_shape by maximum."""
    shape = (new_shape[0], arr.shape[0] // new_shape[0],
             new_shape[1], arr.shape[1] // new_shape[1])
    return arr.reshape(shape).max(-1).max(1)
```

```python
new_shape = (128,128)
newimg = rebin(img, new_shape)
```

```python
plt.imshow(newimg,cmap='binary',origin="lower")
```

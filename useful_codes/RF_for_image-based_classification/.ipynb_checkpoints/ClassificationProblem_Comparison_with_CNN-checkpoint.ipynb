```python
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        cm = cm*100.
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    cm[ np.isnan(cm) ] = 0
    #print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.ylim(-0.5,cm.shape[1]-0.5)
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt)+'%' if fmt=='.2f' else format(cm[i, j], fmt),
                 horizontalalignment="center",verticalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
def calc_phase(t,t0,P):
  return (t-t0)/P - ((t-t0)/P).astype(int)

def sine(time, amp, best_freq, phase):
    y = amp*np.cos(2*np.pi*best_freq*time + phase )
    return y

def fit_all(*arg):
    time = arg[0]
    best_freq = arg[1]
    nparams = (len(arg)-2)//2
    amps = arg[2:2+nparams]
    phases = arg[2+nparams:]
    y = 0
    for i in range(len(amps)):
        y += sine(time, amps[i], (i+1)*best_freq, phases[i])
    return y

def fit_freqs(t,y,nfreq = 3,fix_per=None,plotting = False,ignore_daily_alias=False):
    yres = y.copy()
    
    freqs = []
    amps = []
    phases = []

    for i in range(nfreq):
        if fix_per is None:
            ls = LombScargle(t, yres, nterms=1)
            freq, power = ls.autopower(normalization='psd',minimum_frequency=0.001,maximum_frequency=20, samples_per_peak=10)

        if i == 0:
            if fix_per is None:
                best_freq = freq[power.argmax()]
            else:
                best_freq = 1/fix_per
            #best_power = power[power.argmax()]

            if ignore_daily_alias and  ( np.abs(1-(1/best_freq)) ) < 0.005: #or ( np.abs(0.5-(1/best_freq)) ) < 0.005 ):
                pfit, pcov = curve_fit(lambda time, amp, phase: sine(time, amp, best_freq, phase), t, yres-np.mean(yres),
                                  p0=(np.ptp(yres)/4,3.), bounds=(0, [np.ptp(yres), 2*np.pi]) ,maxfev=5000)

                pfit, pcov = curve_fit(sine, t, yres-np.mean(yres),
                                  p0=(pfit[0],best_freq,pfit[1]), bounds=(0, [np.ptp(yres), 2*best_freq, 2*np.pi]),maxfev=5000 )
                
                yres -= sine(t,pfit[0],pfit[1],pfit[2])
                
                ls = LombScargle(t, yres, nterms=1)
                freq, power = ls.autopower(normalization='psd',
                                                   minimum_frequency=0.001,
                                                   maximum_frequency=20, samples_per_peak=10)
                
                best_freq = freq[power.argmax()]

            pfit, pcov = curve_fit(lambda time, amp, phase: sine(time, amp, best_freq, phase), t, yres-np.mean(yres),
                                  p0=(np.ptp(yres)/4,3.), bounds=(0, [np.ptp(yres), 2*np.pi]) ,maxfev=5000)

            if fix_per is None:
                pfit, pcov = curve_fit(sine, t, yres-np.mean(yres),
                                  p0=(pfit[0],best_freq,pfit[1]), bounds=(0, [np.ptp(yres), 2*best_freq, 2*np.pi]),maxfev=5000 )

                best_freq = pfit[1]
        else:
            pfit, pcov = curve_fit(lambda time, amp, phase: sine(time, amp, (i+1)*best_freq, phase), t, yres-np.mean(yres),
                                  p0=(np.ptp(yres)/4,2.), bounds=(0, [np.ptp(yres), 2*np.pi]),maxfev=5000 )

        if plotting:
            if i==0:
                per = 1/pfit[1]
            else:
                per = 1/( (i+1)*best_freq )
            plt.subplot(121)
            plt.plot(freq, power)
            plt.subplot(122)
            plt.plot(t%per/per,yres-yres.mean(),'.')
            if i == 0:
                plt.plot(t%per/per,sine(t,pfit[0],pfit[1],pfit[2]),'.')
            else:
                plt.plot(t%per/per,sine(t,pfit[0],(i+1)*best_freq,pfit[1]),'.')
            plt.show()

        if i==0 and fix_per is None:
            freqs.append( pfit[1] )
            amps.append( pfit[0] )
            phases.append( pfit[2] )

            yres -= sine(t,pfit[0],pfit[1],pfit[2])
        elif i==0:
            freqs.append( 1/fix_per )
            amps.append( pfit[0] )
            phases.append( pfit[1] )

            yres -= sine(t,pfit[0],1/fix_per,pfit[1])        
        else:
            freqs.append( (i+1)*best_freq )
            amps.append( pfit[0] )
            phases.append( pfit[1] )

            yres -= sine(t,pfit[0],(i+1)*best_freq,pfit[1])
    try: 
        pfit, pcov = curve_fit(fit_all, t, y-np.mean(y),p0=(freqs[0], *amps, *phases),maxfev=5000 )
        return pfit
    except RuntimeError:
        return [freqs[0], *amps, *phases]
    
```

```python
def calc_ecl_fourier(ID,t0,P,which_ogle='III'):
    if which_ogle == 'III':
        t,y,err = np.loadtxt('/Users/batty/Downloads/rozi_variables/OGLEIII/I/%s.dat' % ID,unpack=True)
    elif which_ogle == 'IV':
        t,y,err = np.loadtxt('/Users/batty/Downloads/rozi_variables/OGLEIV/I/%s.dat' % ID,unpack=True)
    else:
        raise ValueError('No such OGLE!')
        
    # Elliminate nan values
    t = t[ np.isfinite(y) ] 
    y = y[ np.isfinite(y) ]

    um = (y < np.median(y)+4*np.std(y)) & (y > np.median(y)-4*np.std(y))

    t = t[um]
    err = err[um]
    y= y[um]

    t0 -= (t0//P)*P + P/2

    P=P/2

    pfit = fit_freqs(t,y,nfreq = 6,fix_per=P,plotting = False,ignore_daily_alias=False)

    #plt.plot(calc_phase(t,t0,P),y-y.mean(),'k.')
    #plt.plot(calc_phase(t,t0,P)+1,y-y.mean(),'k.')
    #plt.plot(np.sort(calc_phase(t,t0,P)),fit_all(t,*pfit)[np.argsort(calc_phase(t,t0,P))])
    #plt.plot(np.sort(calc_phase(t,t0,P))+1,fit_all(t,*pfit)[np.argsort(calc_phase(t,t0,P))])
    #plt.gca().invert_yaxis()
    #plt.xlim(0,1.5)
    #plt.show()

    nfreq = 6 - 3
    period = 1/pfit[0]
    A1 = np.ptp(fit_all(t,*pfit)[np.argsort(calc_phase(t,t0,P))])
    R21 = pfit[2]/pfit[1]
    P21 = (pfit[5+nfreq]-2*pfit[4+nfreq])%(2*np.pi)

    #print(period,A1,R21,P21)
    
    return period,A1,R21,P21
```

### Feature importance

```python
# function for creating a feature importance dataframe
def imp_df(column_names, importances):
    df = pd.DataFrame({'feature': column_names,
                       'feature_importance': importances}) \
           .sort_values('feature_importance', ascending = False) \
           .reset_index(drop = True)
    return df

# plotting a feature importance dataframe (horizontal barchart)
def var_imp_plot(imp_df, title):
    imp_df.columns = ['feature', 'feature_importance']
    sns.barplot(x = 'feature_importance', y = 'feature', data = imp_df, orient = 'h', color = 'royalblue') \
       .set_title(title, fontsize = 20)
```

```python
from sklearn.base import clone 

def drop_col_feat_imp(model, X_train, y_train, random_state = 42):
    
    # clone the model to have the exact same specification as the one initially trained
    model_clone = clone(model)
    # set random_state for comparability
    model_clone.random_state = random_state
    # training and scoring the benchmark model
    model_clone.fit(X_train, y_train)
    benchmark_score = model_clone.score(X_train, y_train)
    # list for storing feature importances
    importances = []
    
    # iterating over all columns and storing feature importance (difference between benchmark and new model)
    for col in X_train.columns:
        model_clone = clone(model)
        model_clone.random_state = random_state
        model_clone.fit(X_train.drop(col, axis = 1), y_train)
        drop_col_score = model_clone.score(X_train.drop(col, axis = 1), y_train)
        importances.append(benchmark_score - drop_col_score)
    
    importances_df = imp_df(X_train.columns, importances)
    return importances_df
```

```python
from sklearn.metrics import r2_score
from rfpimp import permutation_importances

def r2(rf, X_train, y_train):
    return r2_score(y_train, rf.predict(X_train))
```

# Plot the decision boundary of the Random Forest classifier

```python
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap, BoundaryNorm
import matplotlib.patches as mpatches

def plot_rndf(clf,X, y, xlogscale=False, ylogscale=False):
    xlabel = list(X)[0]
    ylabel = list(X)[1]
    
    X_mat = X.as_matrix()
    y_mat = y.as_matrix()

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF', '#AFAFAF', '#AFAAAA'])
    cmap_bold  = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#0F0F0F', '#0F0000'])

    clf.fit(X_mat, y_mat)

    # Plot the decision boundary by assigning a color in the color map
    # to each mesh point.
    
    mesh_step_size = .08  # step size in the mesh
    plot_symbol_size = 30
    
    x_min, x_max = X_mat[:, 0].min() - 1, X_mat[:, 0].max() + 1
    y_min, y_max = X_mat[:, 1].min() - 1, X_mat[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, mesh_step_size),
                         np.arange(y_min, y_max, mesh_step_size))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(15,12))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot training points
    plt.scatter(X_mat[:, 0], X_mat[:, 1], s=plot_symbol_size, c=y, cmap=cmap_bold, edgecolor = 'black')
    plt.xlim(0.01, xx.max())
    plt.ylim(0.01, yy.max())

    #patch0 = mpatches.Patch(color='#FF0000', label='RRab')
    #patch1 = mpatches.Patch(color='#00FF00', label='RRc')
    #patch2 = mpatches.Patch(color='#0000FF', label='CepF')
    #patch3 = mpatches.Patch(color='#AFAFAF', label='Cep1O')
    #patch4 = mpatches.Patch(color='magenta', label='acepF')
    #patch5 = mpatches.Patch(color='yellow', label='acep1O')
    #patch6 = mpatches.Patch(color='brown', label='t2cep')
    #patch7 = mpatches.Patch(color='azure', label='ecl')
    #plt.legend(handles=[patch0, patch1, patch2, patch3, patch4, patch5, patch6, patch7])
    patch0 = mpatches.Patch(color='#FF0000', label='ACep')
    patch1 = mpatches.Patch(color='#00FF00', label='DSct')
    patch2 = mpatches.Patch(color='#0000FF', label='RRL')
    patch3 = mpatches.Patch(color='#0F0F0F', label='T2Cep')
    patch4 = mpatches.Patch(color='#0F0000', label='ECL')
    plt.legend(handles=[patch0, patch1, patch2, patch3,patch4])
        
    plt.xlabel('%s' % xlabel)
    plt.ylabel('%s' % ylabel)
    plt.title("Random Forest classification")
    
    if xlogscale:
        plt.xscale("log")
    if ylogscale:
        plt.yscale("log")
    plt.show()
```

```python
X_plot = input_data[['period','A1']]
y_plot = input_data['label']

X_train_plot, X_test_plot, y_train_plot, y_test_plot = train_test_split(X_plot, y_plot, random_state=0, test_size=0.2)

plot_rndf(rndf,X_train_plot, y_train_plot,xlogscale=True,ylogscale=True)
```

```python
####################
# Testing the OGLE 3 DNN
####################

import numpy as np
import glob

from tensorflow import keras
from tensorflow.keras.preprocessing import image

float_formatter = lambda x: "%.2f" % x
np.set_printoptions(formatter={'float_kind':float_formatter})
```

```python
model = keras.models.load_model('../OGLE3_DNN_results/OGLE3_colored.h5')
# model = Sequential(model)
```

```python
dir = "../testdata_color/"
datafiles = []
datafiles = sorted(glob.glob(dir + "t2cep_*.png"))

# datafiles = ["../testdata_nocolor/acep_001.png"]

print("Number of samples: ", len(datafiles))
```

```python
datafiles[1]
```

```python
prediction = np.array([0,0,0,0,0])

for img in datafiles:
    print(img)
    img_size = 128
    # img = cv2.imread(img)
    # img = cv2.resize(img,(img_size,img_size))

    # img = np.expand_dims(img, axis=0)

    # model.predict(img)
    img = image.load_img(img, target_size=(img_size, img_size))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
#     print(model.predict(img))
    prediction = prediction + model.predict(img)
    # model.predict_classes(img)

print(prediction)
```

```python
prediction/len(datafiles)*100
```
